```python
import xarray as xr
import numpy as np
```

```python
ds = xr.Dataset({
    'var1': (('x','y'), np.random.rand(4,5)),
    'var2': (('x','y'), np.random.rand(4,5))
})
```

```python
#assign some coordinates to the dimensions x,y just for show:
x_values = [1,2,3,4]  #equally could use label values of 'x1', 'x2', 'x3', 'x4' for example
ds = ds.assign_coords(coords=dict
    (
    x=x_values,
    y=['y1','y2','y3','y4','y5']
    )
)  #could have others here too
```

```python
#same thing using curly brackets:
#assign some coordinates to the dimensions x,y just for show:
x_values = [1,2,3,4]  #equally could use label values of 'x1', 'x2', 'x3', 'x4' for example
ds = ds.assign_coords({"x":x_values,
                       "y":['y1','y2','y3','y4','y5']
                      })
```

# Define new dimension values
new_dim_values = ['a', 'b', 'c']

# Add new dimension to coordinates (optional)
ds = ds.assign_coords(new_dim=new_dim_values)

Step 2: Expand Specific Variables to Include the New Dimension

Only the variables that need the new dimension should be expanded.

python

# Expand 'var2' to include 'new_dim'
ds['var2'] = ds['var2'].expand_dims('new_dim')

# Alternatively, expand and assign in one step
ds['var2'] = ds['var2'].expand_dims({'new_dim': new_dim_values})

Now, var2 has the dimensions ('new_dim', 'x', 'y'), while var1 remains with dimensions ('x', 'y').
Step 3: (Optional) Add a New Variable with the New Dimension

If you're adding a new variable that uses the new dimension:

Summary

    Variables Not Expanded: Remain unaffected by the new dimension.
    Variables Expanded: Only these variables include the new dimension.
    Dataset Coordinates: Can include the new dimension without forcing it onto all variables.

By selectively expanding variables, you maintain control over which ones inherit the new dimension.

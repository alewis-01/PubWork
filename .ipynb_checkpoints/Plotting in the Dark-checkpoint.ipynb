#### try to learn how to use plt ...

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x,z)
plt.show()

#y = np.cos(x)

plt.subplot(1,2,1)
plt.plot(x, y)
plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
print('ok')
```

```python
fig = plt.figure(figsize=(2, 2), facecolor='lightskyblue',
                 layout='constrained')
fig.suptitle('Figure')
ax = fig.add_subplot()
ax.set_title('Axes', loc='left', fontstyle='oblique', fontsize='medium')
```

##### Starting with the basics from on-line tutorials

```python
plt.plot([1,2,3,4])
plt.ylabel('some numbers') #uses the index value as x
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16]) #assumes y, x
plt.axis((0,6,0,20)) #sets the axis ranges
plt.ylabel('some numbers')
plt.title('some stuff')
plt.show()
```

```python
# evenly sampled time at 200ms intervals 
t = np.arange(0., 5., 0.2)  #seconds
#plt.plot(t, t, 'r--', t**2)
# red dashes, bluse square and gerren tiangles...
plt.figure(figsize=[3,4]) #must be before the plt.plot...
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g' ) 
plt.show()
```

```python
print(plt.gcf())
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
fig, axs = plt.subplot_mosaic([['A', 'right'], ['B', 'right']],
                              figsize=(4, 3), layout='constrained')

for ax_name, ax in axs.items():
    ax.text(0.5, 0.8, ax_name+':)', ha='center', va='center')
```

```python
#categorical variables
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 5))

#subplot notation seems to be number of rows, number of columns, sequence number...
plt.subplot(231)
plt.bar(names, values)
plt.subplot(232)
plt.scatter(names, values)
plt.subplot(233)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100.

#c  indicates colour, s indicates size...
#keyword data indicates where the data are coming from??
#plt.scatter('a', 'b', c='c', s='d', data=data) #the original...
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
data
```
